# ***Realizar um filtro para analisar somente o ano de 2019***

***Levando em consideração o ano de 2019, calcularei a variação do pib do ano anterior(2018) e criarei uma nova coluna com nome "Variacao*** 

```python
df["Ano"].value_counts()
```

```python
#criar 2 datasets com respectivos ano 2019 , 2018
df_2019 = df[df["Ano"]==2019]
df_2018 = df[df["Ano"]==2018]
#df_2019['Pib_2018'] = df_2018['Produto Interno Bruto, \na preços correntes\n(R$ 1.000)']
df_2019.columns
```

```python
#Escolher somente as variaveis que serão usadas neste projeto, e renomea-las, criarei uma novo dataset.
df_1 = df_2019['Ano']
df_2 = df_2019['Nome do Município']
df_3 = df_2019['Sigla da Unidade da Federação']
df_4 = round(df_2019['Valor adicionado bruto da Agropecuária, \na preços correntes\n(R$ 1.000)'])
df_5 = round(df_2019['Valor adicionado bruto da Indústria,\na preços correntes\n(R$ 1.000)'])
df_6 = round(df_2019['Valor adicionado bruto dos Serviços,\na preços correntes \n- exceto Administração, defesa, educação e saúde públicas e seguridade social\n(R$ 1.000)'])
df_7 = round(df_2019['Valor adicionado bruto da Administração, defesa, educação e saúde públicas e seguridade social, \na preços correntes\n(R$ 1.000)'])
df_8 = round(df_2019['Valor adicionado bruto total, \na preços correntes\n(R$ 1.000)'])
df_9 = round(df_2019['Impostos, líquidos de subsídios, sobre produtos, \na preços correntes\n(R$ 1.000)'])
df_10 = round(df_2019['Produto Interno Bruto, \na preços correntes\n(R$ 1.000)'])
df_2019 = pd.concat([df_1,df_2,df_3,df_4,df_5,df_6,df_7,df_8,df_9,df_10],axis=1)
cols = ['Ano',"Municipio","Estado","Valor_Agro","Valor_Industria","Valor_Servicos","Valor_Admin","Valor_Bruto_Total",
       "Imposto","Pib"]
df_2019.columns = cols
df_2019.head()
```

```python
#calcula da taxa de variacao  -  taxa = ((a - b)/ b) *100
#cria uma coluna "Variacao" e uma coluna "Pib_2018"
df_2019['Variacao'] = ((df_2019['Pib'].values - df_2018['Produto Interno Bruto, \na preços correntes\n(R$ 1.000)'].values) / df_2018['Produto Interno Bruto, \na preços correntes\n(R$ 1.000)'].values) * 100
df_2019['Pib_2018'] = df_2018['Produto Interno Bruto, \na preços correntes\n(R$ 1.000)'].values
df_2019.head()
```

***Considerando a media 1326594(pib), quantos municipios e regioes estão acima da média?***

```python
df_2019['Pib'].mean()
```

```python
#total de municipios
total_municipio = df_2019["Municipio"].count()
total_municipio
```

```python
#media produto interno bruto
cont_media = df_2019['Pib'].mean()
cont_media
```

```python
#somente 659 municipios estão acima da media
cont_pib = df_2019[(df_2019['Pib'] >= 1326594.4371633753 )].count()["Municipio"]
cont_pib
```

```python
#percentual do pib dos municipios que estão acima da media é de 12% arredondando.
percentual_pib = cont_pib/ float(total_municipio) * 100
percentual_pib
```

```python
#calculando por Estado que estao acima da media São Paulo lidera
municipio_estado_media = df_2019[(df_2019['Pib'] >= 1326594.4371633753)]["Estado"].value_counts()
municipio_estado_media
```

# ***As cidades com maior PIB per Capita , top 20!***

```python
# Agrupando os dados por cidades e pib per capita
cidades_pib_top10 = df_2019.groupby(['Municipio'])['Pib','Valor_Agro',"Valor_Industria","Valor_Servicos",'Valor_Admin'].sum().reset_index().rename(columns = {'Pib': 'Total_Pib'})

# Retornamos os Top 20
cidades_pib_top20 = cidades_pib_top10.nlargest(20, "Total_Pib")
cidades_pib_top20
```

```python
#==========Criaçao do grafico================
#tamanho da figura
plt.figure(figsize=(16,7))


#criação do grafico de barras
figura = sns.barplot(x = 'Municipio', y = 'Total_Pib',estimator=np.sum,data=cidades_pib_top20,color='blue')

#label do eixo x
plt.xlabel("Municipios", fontsize=16, color='black')

#label eixo y
plt.ylabel("Pib per Capita", fontsize=16, color='black')

#titulo
plt.title("Top 20 - Municipios com maior PIB per Capita 2019", fontsize=16, color='black')


    
#rotacao eixo
plt.xticks(rotation=70)


plt.show()
```
